```python
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division
%matplotlib inline
```

```python

fig = plt.figure()

X1 = np.arange(0, 1, 0.0025)
X2 = np.arange(0, 1, 0.0025)
X1, X2 = np.meshgrid(X1, X2)
Y = 1+X1+X2

X1 = X1/Y
X2 = X2/Y
X1.reshape((1,-1))
X2.reshape((1,-1))

plt.scatter(X1,X2)
plt.show()
```
