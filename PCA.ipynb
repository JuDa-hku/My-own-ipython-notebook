## Generating Data

To run a naive example for PCA, we generate 200*2 sample from multinomial normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Cov = np.array([[3, 1], [1, 4]])
X = np.random.multivariate_normal([1,2], Cov, size=200)
```

plot these data first

```python
plt.figure(figsize = (4,4))
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
```

```python
print np.cov(X, rowvar=False)
# row will be treated as observations when rowvar=False
```

##Running PCA  
We would like to find the direction in which the data varies most.

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
X_pca = pca.fit_transform(X)
pca.components_
```

```python
pca.mean_
```

```python
plt.figure(figsize=(4,4))
plt.scatter(X_pca[:,0], X_pca[:,1])
plt.axis('equal')
print np.cov(X_pca, rowvar=False)
```

plot the reconstructed data together with the original data

```python
X_rec = pca.inverse_transform(X_transform)
plt.figure(figsize=(10,10))
for i in xrange(4):
    for j in xrange(4):
        plt.subplot(4, 4, i * 4 + j + 1)
        plt.scatter(X[:,i], X[:,j],c='r')
        plt.scatter(X_rec[:,i], X_rec[:,j],c='b')
        plt.axis('equal')
```

# PCA on images

Classic digit handwriting cases.

```python
from sklearn.datasets import fetch_mldata
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
X_digits, _,_, Y_digits = fetch_mldata("MNIST Original").values() # fetch dataset from internet
X_digits, Y_digits = shuffle(X_digits,Y_digits) # shuffle dataset (which is ordered!)
X_digits = X_digits[-5000:]       # take only the last instances, to shorten runtime of PCA

plt.rc("image", cmap="binary")
plt.figure(figsize=(8,4))
for i in xrange(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
X2_digits = pca.fit_transform(X_digits)
##look at these most important directions
plt.figure(figsize=(8,6))
W = pca.components_

for i in xrange(10): # loop over all means
    plt.subplot(2,5,i+1)
    plt.imshow(W[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()

```

```python
plt.plot(pca.explained_variance_);
```

Reconstruct the data again using 28*28 dimensional data

```python
from sklearn.decomposition import PCA
pca = PCA(20)
X2_few_digits = pca.fit_transform(X_digits)
plt.figure(figsize=(16,4))
X_recons_digits = pca.inverse_transform(X2_few_digits)
for i in xrange(10):
    plt.subplot(2,10,i+1)
    plt.imshow(X_recons_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
for i in xrange(10):
    plt.subplot(2,10,11+i)
    plt.imshow(X_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())
plt.tight_layout()
```
