### The curse of high dimensional

The curse of dimensionality refers to various phenomena that arise when analyzing and organizing data in high-dimensional spaces (often with hundreds or thousands of dimensions) that do not occur in low-dimensional settings such as the three-dimensional physical space of everyday experience. [Wiki](http://en.wikipedia.org/wiki/Curse_of_dimensionality)

```python
import numpy as np
import matplotlib.pyplot as plt
def stimulate(d, sampleSize=100):
    reslist = []
    data = np.random.uniform(0,1,d*sampleSize)
    data = data.reshape((sampleSize, d))
    for point in data:
        if sum(np.square(point))<=1:
            reslist.append(1)
        else:
            reslist.append(0)
    return reslist
dims = xrange(1,10)
estimate_volume = []
for dim in dims:
    reslist = stimulate(dim, sampleSize=1000)
    estimate_volume.append(np.mean(reslist))
plt.plot(dims, estimate_volume)
plt.title('Dim vs Circle Area')
```

## What can we do?

In statistics, “curse of dimensionality” is often used to refer to the difficulty of fitting a model when many possible predictors are available. But this expression bothers me, because more predictors is more data, and it should not be a “curse” to have more data. Maybe in practice it’s a curse to have more data (just as, in practice, giving people too much good food can make them fat), but “curse” seems a little strong.

With multilevel modeling, there is no curse of dimensionality. When many measurements are taken on each observation, these measurements can themselves be grouped. Having more measurements in a group gives us more data to estimate group-level parameters (such as the standard deviation of the group effects and also coefficients for group-level predictors, if available).

[From Gelman's blog](http://andrewgelman.com/2004/10/27/the_blessing_of/)
